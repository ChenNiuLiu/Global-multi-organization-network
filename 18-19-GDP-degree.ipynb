```python
import pandas as pd

# Assuming your data is in a DataFrame called 'data'
# Read the CSV file
data = pd.read_csv(r'F:\Gdelt-ergm\18-19(2019-1).csv')

# Assuming your data is in a DataFrame called 'data'
# Replace the 1, 0 with True, False in the respective columns
data['nodeType_IGO'] = data['nodeType_IGO'] == 1
data['nodeType_MNC'] = data['nodeType_MNC'] == 1
data['nodeType_NGO'] = data['nodeType_NGO'] == 1

# Use np.select to create the 'nodeType' column based on the conditions
conditions = [
    data['nodeType_IGO'],
    data['nodeType_MNC'],
    data['nodeType_NGO']
]

choices = [1, 2, 3]

data['nodeType'] = pd.np.select(conditions, choices, default=0)

# If you want to drop the original columns, uncomment the following line
data.drop(columns=['nodeType_IGO', 'nodeType_MNC', 'nodeType_NGO'], inplace=True)

import pandas as pd

# Assuming your data is in a DataFrame called 'data'

# Create the 'nodeTheme' column
themes = [
    "nodeTheme_Agriculture", "nodeTheme_Automobile", "nodeTheme_Aviation", "nodeTheme_Chemical",
    "nodeTheme_Economic", "nodeTheme_Education", "nodeTheme_Energy", "nodeTheme_Enviroment",
    "nodeTheme_Health", "nodeTheme_Human", "nodeTheme_Media", "nodeTheme_Metal",
    "nodeTheme_Militry", "nodeTheme_Politics", "nodeTheme_Service", "nodeTheme_Technology"
]

data["nodeTheme"] = data[themes].idxmax(axis=1)
data["nodeTheme"] = data["nodeTheme"].str.replace("nodeTheme_", "").astype("category").cat.codes + 1

# Create the 'nodeCode' column
codes = [
    "nodeCode_AUS", "nodeCode_AUT", "nodeCode_BEL", "nodeCode_BFA", "nodeCode_BGD", "nodeCode_BLR",
    "nodeCode_BLZ", "nodeCode_BRA", "nodeCode_BWA", "nodeCode_CAF", "nodeCode_CAN", "nodeCode_CHE",
    "nodeCode_CHN", "nodeCode_CMR", "nodeCode_CZE", "nodeCode_DEU", "nodeCode_DJI", "nodeCode_DNK",
    "nodeCode_EGY", "nodeCode_ESP", "nodeCode_ETH", "nodeCode_FIN", "nodeCode_FRA", "nodeCode_GBR",
    "nodeCode_GHA", "nodeCode_GTM", "nodeCode_IDN", "nodeCode_IND", "nodeCode_IRL", "nodeCode_ISR",
    "nodeCode_ITA", "nodeCode_JPN", "nodeCode_KOR", "nodeCode_LBY", "nodeCode_LCA", "nodeCode_MYS",
    "nodeCode_NCL", "nodeCode_NGA", "nodeCode_NLD", "nodeCode_PHL", "nodeCode_RUS", "nodeCode_SAU",
    "nodeCode_SGP", "nodeCode_THA", "nodeCode_TUN", "nodeCode_TWN", "nodeCode_TZA", "nodeCode_USA",
    "nodeCode_VNM", "nodeCode_ZAF", "nodeCode_ZMB"
]

data["nodeCode"] = data[codes].idxmax(axis=1)
data["nodeCode"] = data["nodeCode"].str.replace("nodeCode_", "").astype("category").cat.codes + 1
data=data[['node','nodeGDP','nodeType','nodeTheme','nodeCode']]

# Assuming your data is in a DataFrame called 'data' and has a column named 'nodeGDP'

# Calculate the 25th and 75th percentiles of nodeGDP
q15 = data['nodeGDP'].quantile(0.15)
q50 = data['nodeGDP'].quantile(0.50)

# Create a function to categorize the GDP values
def categorize_gdp(gdp):
    if gdp <= q15:
        return "LowGDP"
    elif gdp <= q50:
        return "MidGDP"
    else:
        return "HighGDP"

# Apply the function to the nodeGDP column and store the results in the GDP_degree column
data['GDP_degree'] = data['nodeGDP'].apply(categorize_gdp)
# Sort the values in the GDP_degree column into categories (1, 2, 3)
# Map the values in the GDP_degree column to categories (1, 2, 3)
data['GDP_degree_sorted'] = data['GDP_degree'].map({'LowGDP': 1, 'MidGDP': 2, 'HighGDP': 3})
data.to_csv(r'F:\Gdelt-ergm\18-19(2019-11).csv')
```
